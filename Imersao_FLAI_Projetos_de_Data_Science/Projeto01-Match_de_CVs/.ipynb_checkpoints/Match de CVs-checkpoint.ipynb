## Pré-processamento de Texto

```python
# word_tokenize() -> quebra o texto em palavras individuais
nltk.download("punkt")

lista_palavras = nltk.tokenize.word_tokenize(texto_CRU)
```

```python
display(lista_palavras)
print()
print("Tamanho da lista:", len(lista_palavras))
```

```python
# Padronizando as palavras em Lowercase (ie, apenas letras minusculas)
lista_palavras = [palavra.lower() for palavra in lista_palavras]
display(lista_palavras)
```

```python
# Pontuacoes e caracteres a serem removidos
lista_remover = ["(',')",";",":","[',']",",", "●", "@", "|", ".", "-", "(", ")", "etc"]
```

```python
# Lista de StopWords - Palavras sem muito valor para agregar
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('portuguese')
display(stop_words)
```

```python
# Lista de palavras sem StopWords e acentuacoes
keywords = [palavra for palavra in lista_palavras if palavra not in stop_words and palavra not in lista_remover]
```

```python
keywords
```

```python
print("Comprimento da lista de palavras filtradas:", len(keywords))
```

```python
# Concatenando as palavras
texto_CV = " ". join(s for s in keywords)
display(texto_CV)
```

### Automatizando nossa Nuvem de palavras

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def NuvemDePalavras(cv, salvar = True):
    '''
    cv: caminho de um arquivo PDF 
    '''
    arq_PDF = pdfplumber.open(cv)
    primeira_pag = arq_PDF.pages[0]  # Le apenas a primeira pagina
    texto_CRU = primeira_pag.extract_text()

    lista_palavras = nltk.tokenize.word_tokenize(texto_CRU)  # Transforma o texto cru em uma lista de termos
    lista_palavras = [palavra.lower() for palavra in lista_palavras]  # Convertendo tudo para minusculo

    keywords = [palavra for palavra in lista_palavras if palavra not in stop_words and palavra not in lista_remover]  # Tira as pontuacoes e stopwords 
    texto_CV = " ".join(s for s in keywords)  # junta tudo em um texto só novamente. 


    wordcloud = WordCloud(background_color = '#0f54c9', 
                          max_font_size = 150, 
                          width = 1280, 
                          height = 720, 
                          colormap= 'Blues').generate(texto_CV) 
 
    # mostrar a imagem final
    fig, ax = plt.subplots(figsize=(16, 9))
    ax.imshow(wordcloud)
    ax.set_axis_off()
    plt.imshow(wordcloud)

    if salvar:
        wordcloud.to_file("wordcloud.png")

    plt.show()
```

### Automatizando o Match de CVs

```python
# Pontuacoes e caracteres a serem removidos
lista_remover = ["(',')",";",":","[',']",",", "●", "@", "|", ".", "-", "(", ")", "etc"]

# Stop Words a serem removidas
stop_words = nltk.corpus.stopwords.words('portuguese')

def MatchCV(cv, vaga, limite = 5):
    '''
    Entradas
      cv: caminho de um arquivo PDF
      vaga: dataset de palavras-chave e pesos
    Saida
      score

    '''

    arq_PDF = pdfplumber.open(cv)
    primeira_pag = arq_PDF.pages[0]  # Lê apenas a primeira página
    texto_CRU = primeira_pag.extract_text()

    lista_palavras = nltk.tokenize.word_tokenize(texto_CRU)  # Transforma o texto cru em uma lista de termos
    lista_palavras = [palavra.lower() for palavra in lista_palavras]  # Transformando tudo para minusculo

    keywords = [palavra for palavra in lista_palavras if palavra not in stop_words and palavra not in lista_remover]  # Tira as pontuacoes e stopwords 
    texto_CV = " ".join(s for s in keywords)  # Junta tudo em um texto soh novamente. 

    pesos = list(vaga['pesos'])
    palavras_chaves = list(vaga['palavras-chave'])

    cont = [texto_CV.count(pc) for pc in palavras_chaves]  # Conta quantas vezes cada termo da vaga aparece no texto do cv

    def aux(x, limite):
        return x if x <= limite else limite

    cont = [aux(i, limite) for i in cont]   # Coloca o limite na contagem de palavras

    pmax = np.sum(np.array(pesos) * limite) 

    score = ((np.array(cont) * pesos).sum() / pmax).round(4)

    return score
```

```python
lista_vagas = vagas

lista_CVs = ["/content/LeonardoCV1.pdf",
             "/content/LeonardoCV2.pdf"]
```

```python
# Lista de Listas, cada lista interna é score de uma pessoa nas vagas
pessoas = [[MatchCV(cv, vaga) for vaga in lista_vagas] for cv in lista_CVs]
pessoas
```

```python
# Filtro para pegar apenas os nomes das pessoas na lista de caminhos
nomes = [cv.split('/')[-1].split('.')[0] for cv in lista_CVs]
```

```python
# Conjunto de dados comos matches das pessoas com as vagas
matchs_df = pd.DataFrame(pessoas, columns = nome_vagas, index = nomes)
```
